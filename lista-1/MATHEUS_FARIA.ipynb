```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Função de pertinência triangular

```python
def trimf(x, a, b, c):
    assert a < b < c, "a < b < c"

    mu = list()
    for x_i in x:
        if x_i < a:
            mu.append(0)
        elif a <= x_i <= b:
            mu.append((x_i-a)/(b-a))
        elif b <= x_i <= c:
            mu.append((c-x_i)/(c-b))
        elif c < x_i:
            mu.append(0)
        else:
            raise x
    
    return mu
```

```python
x = np.arange(0, 10.2, 0.2)
```

```python
plt.figure(figsize=[22, 6])
plt.subplot(1, 2, 1)
y1 = trimf(x, 3, 4, 5)
y2 = trimf(x, 2, 4, 7)
y3 = trimf(x, 1, 4, 9)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.subplot(1, 2, 2)
y1 = trimf(x, 2, 3, 5)
y2 = trimf(x, 3, 4, 7)
y3 = trimf(x, 4, 5, 9)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

# Função pertinência trapezoidal

```python
def trapmf(x, a, b, c, d):

    mu = list()
    for x_i in x:
        if x_i <= a:
            mu.append(0)
        elif a <= x_i <= b:
            mu.append((x_i-a)/(b-a))
        elif b < x_i < c:
            mu.append(1)
        elif c <= x_i <= d:
            mu.append((d-x_i)/(d-c))
        elif x_i > d:
            mu.append(0)
        else:
            raise x
    
    return mu
```

```python
x = np.arange(0, 10.1, 0.1)
```

```python
plt.figure(figsize=[22, 6])

y1 = trapmf(x, 2, 3, 7, 9)
y2 = trapmf(x, 3, 4, 6, 8)
y3 = trapmf(x, 4, 5, 5, 7)
y4 = trapmf(x, 5, 6, 4, 6)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.show()
```

# Função de pertinêcia gaussiana

```python
def gaussmf(x, sigma, c):
    return [np.exp(-0.5*(((x_i-c)/sigma)**2)) for x_i in x]
```

```python
x = np.arange(0, 10.1, 0.1)
```

```python
plt.figure(figsize=[22, 6])
plt.subplot(1, 2, 1)
y1 = gaussmf(x, 0.5, 5)
y2 = gaussmf(x, 1, 5)
y3 = gaussmf(x, 2, 5)
y4 = gaussmf(x, 3, 5)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.subplot(1, 2, 2)
y1 = gaussmf(x, 1, 2)
y2 = gaussmf(x, 1, 4)
y3 = gaussmf(x, 1, 6)
y4 = gaussmf(x, 1, 8)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.show()
```

# Função de pertinência Sino Generalizada

```python
def gbellmf(x, c, sigma, inc):
     return 1 / (1 + np.abs((x-inc)/c)**(2*sigma))
```

```python
x = np.arange(0, 10.1, 0.1)
```

```python
plt.figure(figsize=[22, 6])
plt.subplot(1, 2, 1)
y1 = gbellmf(x, 1, 2, 5)
y2 = gbellmf(x, 2, 4, 5)
y3 = gbellmf(x, 3, 6, 5)
y4 = gbellmf(x, 4, 8, 5)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.subplot(1, 2, 2)
y1 = gbellmf(x, 2, 1, 5)
y2 = gbellmf(x, 2, 2, 5)
y3 = gbellmf(x, 2, 4, 5)
y4 = gbellmf(x, 2, 8, 5)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.show()
```

# Função de pertinência sigmoidal

```python
def sigmf(x, a, c):
    return 1 / (1 + np.exp(-a*(x-c)))
```

```python
x = np.arange(0, 10.1, 0.05)
```

```python
plt.figure(figsize=[22, 6])
plt.subplot(1, 2, 1)
y1 = sigmf(x, -1, 5)
y2 = sigmf(x, -3, 5)
y3 = sigmf(x, 4, 5)
y4 = sigmf(x, 8, 5)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.subplot(1, 2, 2)
y1 = sigmf(x, 5, 2)
y2 = sigmf(x, 5, 4)
y3 = sigmf(x, 5, 6)
y4 = sigmf(x, 5, 8)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)

plt.show()
```
