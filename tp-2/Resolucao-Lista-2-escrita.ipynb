```python
def max_product(x, y):
    z = []
    for x1 in x:
        for y1 in y.T:
            z.append(max(np.multiply(x1, y1)))
    return np.array(z).reshape((x.shape[0], y.shape[1]))
```

```python
def max_min(x, y):
    z = []
    for x1 in x:
        for y1 in y.T:
            z.append(max(np.minimum(x1, y1)))
    return np.array(z).reshape((x.shape[0], y.shape[1]))
```

```python
def y1(x):
    return (2/pi)*x + 1

def y2(x):
    return (-2/pi)*x + 1

def y3(x):
    return (2/pi)*x - 3

plt.plot(x, np.cos(x))
plt.plot(x, y1(x))
plt.plot(x, y2(x))
plt.plot(x, y3(x))
plt.ylim(-1, 1)
plt.show()
```

```python
def trimf(x, a, b, c):
    assert a < b < c, "a < b < c"

    mu = list()
    for x_i in x:
        if x_i < a:
            mu.append(0)
        elif a <= x_i <= b:
            mu.append((x_i-a)/(b-a))
        elif b <= x_i <= c:
            mu.append((c-x_i)/(c-b))
        elif c < x_i:
            mu.append(0)
        else:
            raise x
    
    return np.array(mu)


def calc_line(p1, p2):
    a = (p1[1] - p2[1]) / (p1[0] - p2[0])
    b = p1[1] - a*p1[0]
    return a, b

a, b = calc_line([-pi/2, 0], [pi/2, 1])
print(a*pi, b)

mu_a_1 = trimf(x, -pi, -pi/2, pi/2)
plt.plot(x, mu_a_1, color='orange')

mu_a_2 = trimf(x, -pi/2, pi/2, 3*pi/2)
plt.plot(x, mu_a_2, color='green')

mu_a_3 = trimf(x, pi/2, 3*pi/2, 5*pi/2)
plt.plot(x, mu_a_3, color='red')

plt.show()
```

```python
yy_1 = mu_a_1 * y1(x)
yy_2 = mu_a_2 * y2(x)
yy_3 = mu_a_3 * y3(x)

yy = (yy_1 + yy_2 + yy_3)/(mu_a_1 + mu_a_2 + mu_a_3)

plt.plot(x, np.cos(x))
plt.plot(x, yy)
```
